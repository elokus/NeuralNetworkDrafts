## Data Processing

```python
# Gym
import gym
import gym_anytrading

# Stable baselines
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3 import PPO
# Processing libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("data/Binance_BTCUSDT_1h.csv", skiprows=1)
```

```python
df = df.drop(columns=["Volume USDT", "tradecount", "symbol"], axis=1)
```

```python
df.head()
```

```python
nunix = []

for i in df["unix"]:
    if i < 1000000000000:
        r = i * 1000
    else:
        r = i
    nunix.append(r)
    
df["unix"] = nunix
```

```python
df["date"] = pd.to_datetime(df["unix"],unit="ms")
```

```python
df = df.drop(columns=["unix"], axis=1)
```

```python
df = df.rename(columns={"Volume BTC":"volume"})
```

```python
df.head()
```

```python
df.set_index("date", inplace=True)
df.sort_index()
```

```python
df = df.rename(columns={"date":"Date","open":"Open", "high":"High", "low":"Low", "close":"Close"})
```

```python
df.head()
```

```python
df.to_csv("data/BTC_USDT.csv")
```

```python
df = df.sort_index()
```

```python
df.head()
```

## Build Environment

```python
env = gym.make("stocks-v0", df=df, frame_bound=(48,24000), window_size=48)
```

```python
env.signal_features
```

```python
state = env.reset()
while True:
    action = env.action_space.sample()
    n_state, reward, done, info = env.step(action)
    if done:
        print("info", info)
        break
        
plt.figure(figsize=(150,50))
plt.cla()
env.render_all()
plt.show()
```
